```python
from pepdata import iedb
```

```python
df = iedb.mhc.load_dataframe(human=True)
```

```python
# keep only Class II epitopes
print("Total entries: %d" % len(df))
df_class2 = df[(df["MHC"]["MHC allele class"] == "II") ]
print("Class II: %d" % len(df_class2))

df_class2 = df_class2[df_class2["Epitope"]["Object Type"] == "Linear peptide"]
print("Class II linear peptides: %d" % len(df_class2))

df_class2 = df_class2[~df_class2["Epitope"]["Description"].str.contains(" ")]
print("Class II linear peptides without modifications: %d" % len(df_class2))
```
